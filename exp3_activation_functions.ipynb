```python
import numpy as np
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
x_values=np.linspace(-5,5,100)
y_values=sigmoid(x_values)
```

```python
import matplotlib.pyplot as plt
plt.plot(x_values,y_values,label="Sigmoid")
plt.title("sigmoid activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.legend()
plt.show()
```

```python
#hyperbolic tangent (tanh) activation function
def tanh(x):
  return np.tanh(x)
```

```python
y_values_tanh=tanh(x_values)
```

```python
plt.plot(x_values,y_values_tanh,label="tanh",color="orange")
plt.title("tanh activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.legend()
plt.show()
```

3.Rectified linear unit(relu) Activation function

```python
def relu(x):
  return np.maximum(0,x)
```

```python
y_values_relu=relu(x_values)
```

```python
plt.plot(x_values,y_values_relu,label="ReLU",color="green")
plt.title("Relu activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.show()
```

4 leaky rectified linear unit (leaky relu) activation function

```python
def leaky_relu(x,alpha=0.01):
  return np.where(x>0,x,alpha*x)
```

```python
y_values_leaky_relu=leaky_relu(x_values)
```

```python
plt.plot(x_values,y_values_leaky_relu,label="leaky reLU",color="red")
plt.title("leaky ReLU activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.legend()
plt.show()
```

5 Softmax activation function

```python
def softmax(x):
  exp_x=np.exp(x-np.max(x,axis=-1,keepdims=True))
  return exp_x/np.sum(exp_x,axis=-1,keepdims=True)
```

```python
scores=np.array([2.0,1.0,0.1])
probs=softmax(scores)
print("softmax probabilities:",probs)
```
